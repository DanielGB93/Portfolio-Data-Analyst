TEST DEL CHI CUADRADO
Es una prueba solida para determinar la relación entre dos variables cualitativas utilizando tablas de contingencias
Compara la suma de diferencias al cuadrado, entre lo que estamos esperando.
Una de las debilidades de este test, es que no permite cuantificar cada una de las relaciones entre las variables. Pero sí permite afirmaciones globales como "Fumar esta significativamente relacionado con el cáncer de pulmón".


```python
np.random.seed(0) #Esto funciona para obtener los mismos valores que la clase cuando aplique funciones aleatorias
df = df[df["Origin"].isin(["HOU","ATL","IND"])] #Subset con aquellos casos que tengan un origen con los que tenemos
df = df.sample(frac=1) # reordenar la base de datos
df = df[0:10000] #Sólo seleccionamos 10k filas porque entre más grande es el dataset siempre encontraremos una relación aunque sea pequeña entre las variablaes. Y eso no nos interesa
```

```python
df["BigDelay"] = df["ArrDelay"] > 30 #df["BigDelay"] será igual a True cuando se cumpla que df["ArrDelay"]>30 y False en el caso contrario
observados = pd.crosstab(index = df["BigDelay"], columns = df["Origin"], margins= True) #Creamos una tabla de contingencia
```

```python
observados
```

```python
from scipy.stats import chi2_contingency #Paquete para realizar el test
```

```python
test = chi2_contingency(observados)
test
```

Del resultado anterior: 
El primer valor es el estadístico: La suma de las diferencias al cuadrado
El segundo es el valor "p"
Y finalmente la tabla de valores esperados que es lo que compararemos con la tabla de valores observados, lo cuál hará la función automáticamente

```python
esperados = pd.DataFrame(test[3]) #Crear un dataframe de los valores obtenidos en el test para visualizarla de una manera más cómoda
esperados #Valores teóricos esperados si no hubiese ningún tipo de relación
```

```python
esperados_rel = round(esperados.apply(lambda r: r/len(df)*100, axis=1),2)
observados_rel= round(observados.apply(lambda r: r/len(df)*100, axis=1),2)
```

Si el p-valor< 0.05, hay diferencias significativas: Hay relación entre variables.
Si el p-valor> 0.05, no hay diferencias significativas: No hay relaciones entre variables.
