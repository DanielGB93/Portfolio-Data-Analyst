```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(0)
df = pd.read_csv(r"C:\Users\danie\OneDrive\Documentos\Cursos\python_ds_linkedin\base_datos_2008.csv", nrows = 1000000)
df.sample(frac=1).head(100)
```

Procedemos a crear el grÃ¡fico

```python
plt.scatter(x= df.DayofMonth, y= df.ArrDelay, s= df.Distance)
```

```python
#Ahora agregaremos el parÃ¡metro "alpha" el cual hace referencia a la transparencia de las burbujas
plt.scatter(x= df.DayofMonth, y= df.ArrDelay, s= df.Distance, alpha = 0.3)
```

```python
#Quiero ver los dÃ­as de semana
plt.scatter(x= df.DayofMonth, y= df.ArrDelay, s= df.Distance, c=df.DayOfWeek.isin([6,7]))
plt.title("Retrasos en EEUU")
plt.xlabel("Day of Month")
plt.ylabel("Retraso al llegar")
plt.ylim([0, 1700])
plt.xticks([0, 15, 30])
plt.text(x= 28, y= 1200, s="Mi vuelo")
```
