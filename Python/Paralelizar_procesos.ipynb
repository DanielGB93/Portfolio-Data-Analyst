Cuando paralelizamos procesos le estamos diciendo al nuestro ordenador que solamente utilice algunos o todos los núcleos del procesador para nuestra tarea.
En principio debería acelerar nuestros procesos.
Pero a veces puede tener una contrapartida. Ya que el proceso de paralelizar es costoso al principio, ya que debe distribuir toda la información entre núcleos. Y puede no valer la pena para procesos pequeños y para ordenadores de muchos núcleos.

```python
import pandas as pd
import numpy as np
from joblib import Parallel, delayed

df = pd.read_csv(r"C:\Users\danie\OneDrive\Documentos\Cursos\python_ds_linkedin\base_datos_2008.csv", nrows= 100000)
```

```python
#Creamos un df con las últimas columnas nada más (solo a manera de ejemplo)
df_sub = df[['CarrierDelay','WeatherDelay','NASDelay','SecurityDelay','LateAircraftDelay']]
df_sub.head(10)
```

```python
#crear una función que escoja el retraso máximo de cada fila
def retraso_maximo(fila):
    if not np.isnan(fila).any():
        names = ['CarrierDelay','WeatherDelay','NASDelay','SecurityDelay','LateAircraftDelay']
        return names[fila.index(max(fila))]
    else: 
        return "None"
```

```python
results =[]
for fila in df_sub.values.tolist():
    results.append(retraso_maximo(fila))
```

```python
results
```

Ahora lo intentaremos pero paralelizando el proceso

```python
result = Parallel(n_jobs = 1, backend ="multiprocessing")(map(delayed(retraso_maximo), df_sub.values.tolist()))
```
