```python
import pandas as pd
df= pd.read_csv(r"C:\Users\danie\OneDrive\Documentos\Cursos\python_ds_linkedin\base_datos_2008.csv", nrows = 1e6)
```

```python
#para cambiar el nombre a una columna
df["HoursDelay"]= round(df["ArrDelay"]/60)
df["HoursDelay"].head()
```

```python
df["ArrDelay"].head()
```

```python
#Para eliminar una columna
del(df["HoursDelay"])
```

```python
#Para eliminar varias columnas
df = df.drop(["Month", "Diverted"], axis= 1)
#Debemos igualar df a la función para borrar porque queremos que se guarde en df
```

```python
#Para eliminar una fila
df.drop(0)
```

```python
#Eliminar un rango de filas
df.drop(range(0, 100))
```

```python
#crear un nuevo dataframe con una selección de datos de la tabla actual
#En este caso, será un dataframe con los vuelos de Atlanta y otro con los vuelos de Houston
dfATL= df[df.Origin == "ATL"]
dfHOU= df[df.Origin =="HOU"]
```

```python
#podemos combinar los dos df creados
newdf = dfATL.append(dfHOU)
newdf.Origin
```

```python
#GROUPBY
#Sintaxis: Dataframe.groupby(by = "columna")["Columna que quiero ver"] y después puedo agregar cualquier otro parámetro como .max()
df.groupby(by ="DayOfWeek")["ArrDelay"].max()
```

```python
df.groupby(by ="DayOfWeek")["ArrDelay"].describe()
```

```python
df.groupby(by ="DayOfWeek")["ArrDelay","DepDelay"].mean()
```

```python
df.groupby(by ="DayOfWeek")[("ArrDelay", "DepDelay")].max()
```

```python
#Puedo realizar operaciones con estas dos columnas pero tengo que tener cuidado de que sean del mismo tipo de datos
df.groupby(by ="DayOfWeek")["ArrDelay","DepDelay"].max() - df.groupby(by ="DayOfWeek")["ArrDelay","DepDelay"].min()
```

```python
#Crear un DF donde el origen es ATL y HOU
dfATLHOU = df[df.Origin.isin(["ATL", "HOU"])]
```

```python
#Agrupar por el día de la semana y el origen en segundo lugar. Mostrarme los ArrDelay promedio por cada uno
dfATLHOU.groupby(by = ["DayOfWeek", "Origin"])["ArrDelay"].mean()
```

```python
#Si cambo el orden de los parámetros el resultado cambia
dfATLHOU.groupby(by = ["Origin","DayOfWeek"])["ArrDelay"].mean()
```

```python
#Tratar duplicados y perdidos
#Crearemos un df con valores duplicados
dfduplicate = df.append(df)
#Lo ordenamos
dfduplicate= dfduplicate.sample(frac=1)


```

```python
#Borramos las filas duplicadas y lo guardamos en un nuevo dataframe
dfclean = dfduplicate.drop_duplicates()
```

```python
#Verificar que se hayan borrado los datos duplicados
len(df) == len(dfclean)
```

```python
print(len(df))
print(len(dfclean))
```

```python
#Buscar valores duplicados en una columna en específico para borrarlos
dfclean.drop_duplicates(subset = "DayofMonth")
```
