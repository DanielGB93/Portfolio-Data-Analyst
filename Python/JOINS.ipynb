```python
import pandas as pd
#Filas
consumidores = [("A","Móvil"),("B","Móvil"),("A","Portátil"),("A","Tablet"),("B","Tablet"),("C","Portátil"),("D","Smartwatch"),("E","Consola")]
#Columnas
con_labels=["consumidor","Producto"]                                                                                                        
                                                             
                                                        
```

```python
con_df= pd.DataFrame.from_records(consumidores, columns = con_labels)
con_df
```

```python
productores = [("a","Móvil"),("a","Smartwatch"),("a","Tablet"),("b","Portátil"),("c","Sobremesa"),("c","Portátil")]
prod_labels = ["Productor","Producto"]
prod_df = pd.DataFrame.from_records(productores, columns = prod_labels)
prod_df
```

```python
#la función merge del paquete pandas nos permite unir dos dataframes. El parámetro "on" nos indica la columna mediante la cual se unirán
#Los dos DF. El parámetro "how" nos indica el tipo de JOIN que haremos
pd.merge(con_df, prod_df, on = "Producto", how="outer")
```

```python
#El inner Join nos muestra solo los datos que están en las dos tablas
pd.merge(con_df, prod_df, on = "Producto", how="inner")
```

Ya no aparecen los NaN 

```python
#Juntar los dos dataframes sólo si los valores aparecen en el de la derecha (prod_df)
pd.merge(con_df, prod_df, on = "Producto", how="right")
```

```python
#Juntar los dos dataframes sólo si los valores aparecen en el de la izquierda (con_df)
pd.merge(con_df, prod_df, on = "Producto", how="left")
```
